```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
df_players_injuries = pd.read_parquet("s3://tech-challenge-3-curated/PlayersInjuries/")
```

```python
df_players_injuries = df_players_injuries[~df_players_injuries["days"].str.contains("\?", na=False)]
df_players_injuries = df_players_injuries[~df_players_injuries["days"].isnull()]
df_players_injuries["gamesmissed"] = df_players_injuries["gamesmissed"].copy().fillna(0)
df_players_injuries["days"] = df_players_injuries["days"].str.replace(" days", "").astype(int)
df_players_injuries = df_players_injuries[["player_id", "days", "gamesmissed", "injury"]].reset_index(drop=True)
```

```python
set(df_players_injuries["injury"])
```

```python
df_players_injuries["injury"] = df_players_injuries["injury"].copy().str.replace(" ", "_").str.lower()
```

```python
df_players_injuries
```

```python
df_players_injuries.dtypes
```

```python
df_players_injuries["gamesmissed"] = df_players_injuries["gamesmissed"].astype(int)
```

```python
df_players_injuries = df_players_injuries.pivot_table(
    values=["days", "gamesmissed"], index=["player_id"], columns=["injury"], aggfunc="sum", fill_value=0
)
```

```python
df_players_injuries = df_players_injuries.reset_index()
```

```python
df_players_injuries.columns
```

```python
df_players_injuries.columns = [col[0] + "_" + col[1] for col in df_players_injuries.columns]
```

```python
df_players_injuries.rename(columns={"player_id_": "player_id"}, inplace=True)
```
